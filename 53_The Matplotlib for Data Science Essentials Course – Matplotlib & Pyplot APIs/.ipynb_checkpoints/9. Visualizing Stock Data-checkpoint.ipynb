```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
stock_data = pd.read_csv('datasets/stocks.csv')
stock_data.head()
```

```python
stock_data["Date"] = pd.to_datetime(stock_data["Date"])
stock_data.head()
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.plot(stock_data["Date"], stock_data["AAPL"], color="green")
ax1.set_title("Apple VS. IBM")
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.plot(stock_data["Date"], stock_data["AAPL"], color="green")
ax1.set_title("Apple VS. IBM")

ax2.plot(stock_data["Date"], stock_data['IBM'], color="b")
```

```python
fig = plt.figure(figsize=(10, 10))
fig.suptitle("Stock Price Comparison 2007-2017")

# Microsoft Stock Movements
ax1 = fig.add_subplot(221)
ax1.set_title("Microsoft")
ax1.plot(stock_data["Date"], stock_data["MSFT"], color='green')
```

```python
fig = plt.figure(figsize=(10, 10))
fig.suptitle("Stock Price Comparison 2007-2017")

# Microsoft Stock Movements
ax1 = fig.add_subplot(221)
ax1.set_title("Microsoft")
ax1.plot(stock_data["Date"], stock_data["MSFT"], color='green')


# Google Stock Movements
ax1 = fig.add_subplot(222)
ax1.set_title("Google")
ax1.plot(stock_data["Date"], stock_data["GOOG"], color='purple')
```

```python
fig = plt.figure(figsize=(10, 10))
fig.suptitle("Stock Price Comparison 2007-2017")

# Microsoft Stock Movements
ax1 = fig.add_subplot(221)
ax1.set_title("Microsoft")
ax1.plot(stock_data["Date"], stock_data["MSFT"], color='green')


# Google Stock Movements
ax1 = fig.add_subplot(222)
ax1.set_title("Google")
ax1.plot(stock_data["Date"], stock_data["GOOG"], color='purple')


# Statbucks Stock Movements
ax1 = fig.add_subplot(223)
ax1.set_title("StarBucks")
ax1.plot(stock_data["Date"], stock_data["SBUX"], color='y')
```
