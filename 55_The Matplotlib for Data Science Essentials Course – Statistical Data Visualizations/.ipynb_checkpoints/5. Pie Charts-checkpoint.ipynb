```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
data = pd.read_csv('datasets/sector_weighting.csv')
data
```

```python
plt.pie(data['Percentage'], labels=data['Sector'])
```

```python
plt.pie(data['Percentage'], labels=data['Sector'])
plt.axis("equal")
```

```python
colors = ['lawngreen', 'orange', 'teal', 'yellow','violet']
```

```python
plt.pie(data['Percentage'], labels=data['Sector'], colors=colors, autopct="%.2f")
plt.axis("equal")
```

```python
plt.pie(data['Percentage'], labels=data['Sector'], colors=colors, autopct="%.2f", startangle=90, counterclock=False)
plt.axis("equal")
```

```python
explode = (0, 0.1, 0, 0.3, 0)
```

```python
plt.pie(data['Percentage'],
        labels=data['Sector'],
        colors=colors, 
        autopct="%.2f",
        explode=explode)

plt.axis("equal")
```

```python
wedges, texts, autotexts = plt.pie(data['Percentage'],
        labels=data['Sector'],
        colors=colors, 
        autopct="%.2f",
        explode=explode)

plt.axis("equal")

print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

```python
wedges, texts, autotexts = plt.pie(data['Percentage'],
        labels=data['Sector'],
        colors=colors, 
        autopct="%.2f",
        explode=explode)

plt.axis("equal")

wedges[1].set(edgecolor='purple', linewidth=2)

texts[1].set(family='monospace', size=20)

autotexts[1].set(weight='bold', size=15)
```
