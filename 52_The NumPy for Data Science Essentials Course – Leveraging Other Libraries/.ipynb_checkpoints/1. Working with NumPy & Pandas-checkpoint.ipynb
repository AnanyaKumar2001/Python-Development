```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
titanic.head()
```

```python
titanic = titanic.drop(columns=["Cabin"])
```

```python
titanic.head()
```

```python
np.sum(titanic['Age'].isna())
```

```python
port = titanic.groupby('Embarked').aggregate(np.mean)
```

```python
port
```

```python
port[["Age", "Fare"]]
```

```python
total = np.count_nonzero(titanic['PassengerId'])
```

```python
survived = np.count_nonzero(titanic['Survived'])
```

```python
percent_survived = (survived / total) * 100
```

```python
print("Total Number of Passengers = ", total)
print("Passengers who Survived = ", survived)
print("Survival Percentage = ", percent_survived)
```

```python
females = titanic[titanic["Sex"] == "female"]
females
```

```python
females_survived = np.count_nonzero(females["Survived"])
```

```python
print("Number Females who Survived = ", females_survived)
```

```python
males = titanic[titanic["Sex"] == "male"]
males_survived = np.count_nonzero(males["Survived"])
```

```python
print("Number Males who Survived = ", males_survived)
```
