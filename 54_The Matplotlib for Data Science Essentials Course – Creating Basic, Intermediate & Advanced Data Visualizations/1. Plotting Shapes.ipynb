```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
fig, ax = plt.subplots()

ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, fill=False))
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect="equal")

ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, fill=True))
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect="equal")

ax.add_patch(patches.Rectangle((0.1, 0.1), 0.5, 0.5, facecolor='yellow', edgecolor="green"))
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect="equal")

polygon = patches.Polygon([
    [0.1, 0.1],
    [0.2, 0.8],
    [0.5, 0.7],
    [0.8, 0.1],
    [0.4, 0.3]
], fill=False)

ax.add_patch(polygon)
```

```python
fig, ax = plt.subplots()

ax.set_aspect(aspect="equal")

polygon = patches.Polygon([
    [0.1, 0.1],
    [0.2, 0.8],
    [0.5, 0.7],
    [0.8, 0.1],
    [0.4, 0.3]
], fill=False, closed=False)

ax.add_patch(polygon)
```
