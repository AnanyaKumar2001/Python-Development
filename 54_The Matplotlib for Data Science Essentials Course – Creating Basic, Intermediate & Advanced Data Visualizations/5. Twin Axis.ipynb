```python
fig, ax_tempF = plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel("Date")
ax_tempF.tick_params(axis="x", bottom=False, labelbottom=False)

ax_tempF.set_ylabel("Temperature (F)", color="blue", size="x-large")
ax_tempF.tick_params(axis="y", labelcolor="blue", labelsize="large")

ax_tempF.plot(austin_weather['Date'], austin_weather['TempAvgF'], color="blue")

ax_tempC = ax_tempF.twinx()

ymin, ymax = ax_tempF.get_ylim()

ax_tempC.set_ylim(fahrenheit_2_celcius(ymin), fahrenheit_2_celcius(ymax))

ax_tempC.set_ylabel("Temperature (C)", color="purple", size="x-large")
ax_tempC.tick_params(axis="y", labelcolor="purple", labelsize="large")
```
