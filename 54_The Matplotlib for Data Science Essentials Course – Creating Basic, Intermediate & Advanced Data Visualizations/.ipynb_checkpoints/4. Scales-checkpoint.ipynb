```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
np.random.seed()

y=np.random.uniform(low=0.0, high=1000, size=(1000, ))
y.sort()

x = np.arange(len(y))
```

```python
plt.plot(x, y)

plt.grid(True)

plt.yscale('linear')
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log', base=2)
```

```python
plt.plot(x, y)
plt.grid(True)

plt.yscale('log')
plt.xscale('log', base=2)
```
