```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", xy=(1, 2), xytext=(1.5, 2.0), arrowprops=dict(color="g"))
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", xy=(1, 2), xytext=(1, 3), arrowprops=dict(color="g"))
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", xy=(1, 2), xytext=(1, 3), arrowprops=dict(facecolor="y", edgecolor="g", alpha=0.8))
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", 
            xy=(1, 2), 
            xytext=(1, 3), 
            arrowprops=dict(facecolor="y", edgecolor="g", alpha=0.8), 
            color="orange", style="italic"
           )
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", 
            xy=(2, 4), 
            xytext=(2.0, 2.5), 
            arrowprops=dict(facecolor="g", alpha=0.8), 
            color="orange", style="italic"
           )

ax.plot([2], [4], 'ro')
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3],
        [2, 4, 6])

ax.annotate("Min Value", 
            xy=(2, 4), 
            xytext=(2.0, 2.5), 
            arrowprops=dict(facecolor="g", alpha=0.8, shrink=0.1), 
            color="orange", style="italic"
           )

ax.plot([2], [4], 'go')
```

```python
# Advanced Annotations

import numpy as np
```

```python
np.random.seed(0)

# skew towards the bottom left of the plot
x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

# Skew towards the top right of the plot
x2 = 1 + np.random.randn(100)
y2 = 1 + np.random.randn(100)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

bbox_props = dict(boxstyle="square", facecolor="w", alpha=0.5)

ax.text(-2, -2, "Sample A", ha="center", va="center", size=20, bbox=bbox_props)
ax.text(2, 2, "Sample B", ha="center", va="center", size=20, bbox=bbox_props)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

bbox_props = dict(boxstyle="square", facecolor="w", alpha=0.5)

ax.text(-2, -2, "Sample A", ha="center", va="center", size=20, bbox=bbox_props)
ax.text(2, 2, "Sample B", ha="center", va="center", size=20, bbox=bbox_props)

arrow_bbox_props = dict(boxstyle="rarrow", facecolor="#EBF5FB", edgecolor="b", linewidth=2, alpha=0.5)

ax.text(0, 0, "Data Direction", ha="center", va="center", rotation=45, size=12, bbox=arrow_bbox_props)
```

```python
fig, ax = plt.subplots()

ax.scatter(x1, y1, color='r')
ax.scatter(x2, y2, color='g')

sawtooth_bbox_props = dict(boxstyle="sawtooth", facecolor="#ABEBC6", alpha=0.8)

round4_bbox_props = dict(boxstyle="round4", facecolor="#EBDEF0", alpha=0.8)

arrow_bbox_props = dict(boxstyle="darrow", facecolor="#FDEDEC", alpha=0.5)

ax.text(-2, -2, "Sample A", ha="center", va="center", size=20, bbox=sawtooth_bbox_props)
ax.text(2, 2, "Sample B", ha="center", va="center", size=20, bbox=round4_bbox_props)
ax.text(0, 0, "Data Direction", ha="center", va="center", rotation=45, size=12, bbox=arrow_bbox_props)
```
