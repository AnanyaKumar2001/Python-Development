```python
import httplib2
import webbrowser

from pprint import pprint
```

```python
bin_url = "https://httpbin.org/"
```

```python
webbrowser.open(bin_url)
```

```python
http = httplib2.Http()
```

```python
resp, data = http.request(bin_url)
```

```python
pprint(resp)
```

```python
type(resp), len(resp)
```

```python
resp.status, resp.reason, resp.version
```

```python
resp.previous
```

```python
pprint(data)
```

```python
type(data), len(data)
```
